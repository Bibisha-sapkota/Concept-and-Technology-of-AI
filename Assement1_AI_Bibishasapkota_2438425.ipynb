#Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
q1 = filtered_south_asia['score'].quantile(0.25)
q2 = filtered_south_asia['score'].quantile(0.50)  # Median (second quartile)
q3 = filtered_south_asia['score'].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = filtered_south_asia[(filtered_south_asia['score'] < lower_bound) | (filtered_south_asia['score'] > upper_bound)]
print(outliers[['Country name', 'score']])

q1_gdp = filtered_south_asia['Log GDP per capita'].quantile(0.25)
q2_gdp = filtered_south_asia['Log GDP per capita'].quantile(0.50)
q3_gdp = filtered_south_asia['Log GDP per capita'].quantile(0.75)

iqr_gdp = q3_gdp - q1_gdp
lower_bound_gdp = q1_gdp - 1.5 * iqr_gdp
upper_bound_gdp = q3_gdp + 1.5 * iqr_gdp

outliers_gdp = filtered_south_asia[(filtered_south_asia['Log GDP per capita'] < lower_bound_gdp) | (filtered_south_asia['Log GDP per capita'] > upper_bound_gdp)]
print(outliers_gdp[['Country name', 'Log GDP per capita']])

```

2.@@@@ Define outliers using the 1.5 × IQR rule.

3.Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python

q1 = filtered_south_asia['score'].quantile(0.25)
q3 = filtered_south_asia['score'].quantile(0.75)
q1_gdp = filtered_south_asia['Log GDP per capita'].quantile(0.25)
q3_gdp = filtered_south_asia['Log GDP per capita'].quantile(0.75)

iqr_score = q3 - q1
iqr_gdp = q3_gdp - q1_gdp

lower_bound = q1 - 1.5 * iqr_score
upper_bound = q3 + 1.5 * iqr_score
lower_bound_gdp = q1_gdp - 1.5 * iqr_gdp
upper_bound_gdp = q3_gdp + 1.5 * iqr_gdp


outliers = filtered_south_asia[     # Identify outliers based on Score and GDP per Capita
    (filtered_south_asia['score'] < lower_bound) | (filtered_south_asia['score'] > upper_bound) |
    (filtered_south_asia['Log GDP per capita'] < lower_bound_gdp) | (filtered_south_asia['Log GDP per capita'] > upper_bound_gdp)]

plt.figure(figsize=(10, 6))
plt.scatter(filtered_south_asia['Log GDP per capita'], filtered_south_asia['score'], color='blue', label='Non-Outliers')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot with Outliers Highlighted')
plt.legend()
plt.show()
```
